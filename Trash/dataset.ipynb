```python
import numpy as np
import pandas as pd
import json
import os
```

```python
dir="C:/Users/princ/OneDrive/Desktop/AI Project/intents"
questions=[]
tags=[]
files=os.listdir(dir)
for file in files:
    with open(os.path.join(dir,file)) as f:
        data=json.load(f)
        intents=data['intents']
        for intent in intents:
            for pattern in intent['patterns']:
                questions.append(pattern)
                tags.append(intent['tag'])
df=pd.DataFrame({'questions':questions,'tags':tags})
df.to_csv('dataset.csv',index=False)
```
