```python
diseases=['common cold','flu','food allergy','migraine','urinary tract infection','seasonal allergies (hay fever)','tonsillitis','dermatitis due to sun exposure','heat stroke','heat exhaustion','ear wax impaction','drug reaction','drug poisoning due to medication','conjunctivitis','anemia','muscle spasm','cold sore','yeast infection','high blood pressure','chicken pox','dengue fever','malaria','typhoid disease','frostbite']
```

```python
import nltk
```

```python
import json
from nltk.stem import PorterStemmer
stemmer=PorterStemmer()
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense,Input
import numpy as np
```

```python
for disease in diseases:
    disease=disease.lower()
    disease=disease.replace(' ','')
    with open(f'C:\\Users\\princ\\OneDrive\\Desktop\\AI Project\\intents\\intent{disease}.json') as f:
        data=json.load(f)
    data=data['intents']
    words=[]
    docs_x=[]
    labels=[]
    docs_y=[]
    for doc in data:
        labels.append(doc['tag'])
        for pattern in doc['patterns']:
            token=nltk.word_tokenize(pattern)
            wrds=[stemmer.stem(w.lower()) for w in token]
            words.extend(wrds)
            docs_x.append(pattern)
            docs_y.append(doc['tag'])
    words=sorted(list(set(words)))
    training=[]
    output=[]
    temp=[0 for _ in range(len(labels))]
    for i,pattern in enumerate(docs_x):
        ex=[]
        token=nltk.word_tokenize(pattern)
        wrds=[stemmer.stem(w.lower()) for w in token]
        for word in words:
            if word in wrds:
                ex.append(1)
            else:
                ex.append(0)
        index=labels.index(docs_y[i])
        output_i=temp[:]
        output_i[index]=1
        training.append(ex)
        output.append(output_i)
    training=np.array(training)
    output=np.array(output)
    model=Sequential(
        [
            Input(shape=(len(training[0]),)),
            Dense(50,activation='relu'),
            Dense(50,activation='relu'),
            Dense(len(labels),activation='softmax')
        ]
    )
    model.compile(loss='categorical_crossentropy',optimizer='adam',metrics=['accuracy'])
    model.fit(training,output,epochs=100)
    model.save(f'C:\\Users\\princ\\OneDrive\\Desktop\\AI Project\\models\\model{disease}.keras')
```
