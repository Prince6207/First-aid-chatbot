```python
diseases=['common cold','flu','food allergy','migraine','urinary tract infection','seasonal allergies (hay fever)','tonsillitis','dermatitis due to sun exposure','heat stroke','heat exhaustion','ear wax impaction','drug reaction','drug poisoning due to medication','conjunctivitis','anemia','muscle spasm','cold sore','yeast infection','high blood pressure','chicken pox','dengue fever','malaria','typhoid disease','frostbite']
```

```python
import nltk
```

```python
import json
from nltk.stem import PorterStemmer
stemmer=PorterStemmer()
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense,Input
import numpy as np
```

```python
for disease in diseases:
    disease=disease.lower()
    disease=disease.replace(' ','')
    with open(f'C:\\Users\\princ\\OneDrive\\Desktop\\AI Project\\intents\\intent{disease}.json') as f:
        data=json.load(f)
    data=data['intents']
    words=[]
    docs_x=[]
    labels=[]
    docs_y=[]
    for doc in data:
        labels.append(doc['tag'])
        for pattern in doc['patterns']:
            token=nltk.word_tokenize(pattern)
            wrds=[stemmer.stem(w.lower()) for w in token]
            words.extend(wrds)
            docs_x.append(pattern)
            docs_y.append(doc['tag'])
    words=sorted(list(set(words)))
    training=[]
    output=[]
    temp=[0 for _ in range(len(labels))]
    for i,pattern in enumerate(docs_x):
        ex=[]
        token=nltk.word_tokenize(pattern)
        wrds=[stemmer.stem(w.lower()) for w in token]
        for word in words:
            if word in wrds:
                ex.append(1)
            else:
                ex.append(0)
        index=labels.index(docs_y[i])
        output_i=temp[:]
        output_i[index]=1
        training.append(ex)
        output.append(output_i)
    training=np.array(training)
    output=np.array(output)
    model=Sequential(
        [
            Input(shape=(len(training[0]),)),
            Dense(50,activation='relu'),
            Dense(50,activation='relu'),
            Dense(len(labels),activation='softmax')
        ]
    )
    model.compile(loss='categorical_crossentropy',optimizer='adam',metrics=['accuracy'])
    model.fit(training,output,epochs=100)
    model.save(f'C:\\Users\\princ\\OneDrive\\Desktop\\AI Project\\models2\\model{disease}.h5')
```

```python
import numpy as np
import json
import nltk
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, RNN, SimpleRNN, Embedding, LSTM, Input
from tensorflow.keras.preprocessing.text import Tokenizer
from tensorflow.keras.preprocessing.sequence import pad_sequences

# ... (assuming nltk downloads and stemmer are initialized)

for disease in diseases:
    disease_clean = disease.lower().replace(' ', '')
    path = f'C:\\Users\\princ\\OneDrive\\Desktop\\AI Project\\intents\\intent{disease_clean}.json'
    
    with open(path) as f:
        data = json.load(f)['intents']

    patterns = []
    labels = []
    unique_labels = []

    for doc in data:
        tag = doc['tag']
        if tag not in unique_labels:
            unique_labels.append(tag)
        for pattern in doc['patterns']:
            patterns.append(pattern.lower())
            labels.append(tag)

    # 1. Tokenize the text (converts words to unique integers)
    tokenizer = Tokenizer()
    tokenizer.fit_on_texts(patterns)
    sequences = tokenizer.texts_to_sequences(patterns)
    word_index = tokenizer.word_index
    vocab_size = len(word_index) + 1

    # 2. Pad sequences so they are all the same length
    max_len = max([len(x) for x in sequences])
    training_x = pad_sequences(sequences, maxlen=max_len, padding='post')

    # 3. Encode labels (One-Hot)
    label_map = {label: i for i, label in enumerate(unique_labels)}
    training_y = []
    for l in labels:
        target = np.zeros(len(unique_labels))
        target[label_map[l]] = 1
        training_y.append(target)
    
    training_y = np.array(training_y)

    # 4. Build the RNN Model
    model = Sequential([
        Input(shape=(max_len,)),
        # Embedding turns integers into dense vectors of fixed size (e.g., 16)
        Embedding(input_dim=vocab_size, output_dim=16),
        # SimpleRNN or LSTM for sequence processing
        LSTM(32, dropout=0.2), 
        Dense(32, activation='relu'),
        Dense(len(unique_labels), activation='softmax')
    ])

    model.compile(loss='categorical_crossentropy', optimizer='adam', metrics=['accuracy'])
    
    # 5. Train and Save
    model.fit(training_x, training_y, epochs=200, batch_size=5, verbose=1)
    model.save(f'C:\\Users\\princ\\OneDrive\\Desktop\\AI Project\\models3\\model{disease_clean}.h5')
```
